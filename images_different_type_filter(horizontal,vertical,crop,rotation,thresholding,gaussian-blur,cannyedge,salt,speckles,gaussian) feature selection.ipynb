## thresholding

```python
image5 = pyplot.imread('F:\\IACSD-PROJECT-MECHANICAL-IACSD-PROJECT-MECHNICAL-TOOL\\machine-tool-dataset\\train_data\\train_data\\Gasoline Can\\22.jpg')
print(image5.shape)
pyplot.imshow(image5)
pyplot.title('original image'),plt.xticks([]), plt.yticks([])
```

```python
def thres_seg(image5):
    gray = rgb2gray(image5)
    print(gray.shape)
    pyplot.figure() # To create new frame for the image to be displayed
    pyplot.imshow(gray, cmap='gray')
    gray_r = gray.reshape(gray.shape[0]*gray.shape[1])
    gray_r [ gray_r > gray_r.mean() ] = 1
    gray_r [ gray_r <= gray_r.mean() ] = 0
    gray_r = gray_r.reshape(gray.shape[0],gray.shape[1])
    pyplot.figure()
    pyplot.imshow(gray_r, cmap='gray')
```

```python
thres_seg(image5)
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
image11 = cv2.imread('F:\\IACSD-PROJECT-MECHANICAL-IACSD-PROJECT-MECHNICAL-TOOL\\machine-tool-dataset\\train_data\\train_data\\pebbel\\2156.jpg')
plt.imshow(cv2.cvtColor(image11,cv2.COLOR_RGB2BGR))
blue_min = np.array([10, 60, 60], np.uint8)
blue_max = np.array([255, 255, 255], np.uint8)
```

```python
mask_inverse = cv2.inRange(image11, blue_min, blue_max)
mask = cv2.bitwise_not(mask_inverse)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB))
```

```python
mask_rgb = cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)

# perform bitwise and on mask to obtain cut-out image that is not blue
masked_upstate = cv2.bitwise_and(image11, mask_rgb)

# replace the cut-out parts with white
masked_replace_white = cv2.addWeighted(masked_upstate, 1, \
                                       cv2.cvtColor(mask_inverse, cv2.COLOR_GRAY2RGB), 1, 0)

plt.imshow(cv2.cvtColor(masked_replace_white, cv2.COLOR_BGR2RGB))
```

## salt and pepper noise in toolbox image

```python

# Load the image
img12 = cv2.imread("F:\\IACSD-PROJECT-MECHANICAL-IACSD-PROJECT-MECHNICAL-TOOL\\machine-tool-dataset\\train_data\\train_data\\Tool box\\4664.jpg")
 
# Add salt-and-pepper noise to the image.
noise_img = random_noise(img12, mode='s&p',amount=0.3)
 
# The above function returns a floating-point image
# on the range [0, 1], thus we changed it to 'uint8'
# and from [0,255]
noise_img = np.array(255*noise_img, dtype = 'uint8')
 
# Display the noise image
cv2.imshow('salt and pepper noise image',noise_img)
cv2.waitKey(0)
```

## speckles noise

```python
img13 = cv2.imread('F:\\IACSD-PROJECT-MECHANICAL-IACSD-PROJECT-MECHNICAL-TOOL\\machine-tool-dataset\\train_data\\train_data\\Tool box\\4664.jpg')
 
gauss = np.random.normal(0,1,img13.size)
gauss = gauss.reshape(img13.shape[0],img13.shape[1],img13.shape[2]).astype('uint8')
noise = img13 + img13 * gauss
 
cv2.imshow('Speckle Noise',noise)
cv2.waitKey(0)
```

## gaussian noise

```python
img14 = cv2.imread('F:\\IACSD-PROJECT-MECHANICAL-IACSD-PROJECT-MECHNICAL-TOOL\\machine-tool-dataset\\train_data\\train_data\\Tool box\\4678.jpg')
# Generate Gaussian noise
gauss1 = np.random.normal(0,1,img14.size)
gauss2 = gauss1.reshape(img14.shape[0],img14.shape[1],img14.shape[2]).astype('uint8')
# Add the Gaussian noise to the image
img_gauss = cv2.add(img14,gauss2)
# Display the image
cv2.imshow('gaussian noise',img_gauss)
cv2.waitKey(0)
```
